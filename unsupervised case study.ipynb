# Checking and removing outliers in 'fresh' feature

```python
plt.boxplot(df['Fresh'])
plt.title('Fresh')
```

removing outliers using IQR

```python
Q1=np.percentile(df['Fresh'],25,interpolation='midpoint')
Q2=np.percentile(df['Fresh'],50,interpolation='midpoint')   
Q3=np.percentile(df['Fresh'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3) 

#finding IQR
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print(low_lim)          
print(up_lim)
```

```python
#removing outlier using for loop
outlier=[]
for x in df['Fresh']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
        
len(outlier)
```

```python
#Removing the outliers
ind=df['Fresh']>up_lim
df.loc[ind].index
df.drop(df.loc[ind].index,inplace=True)
plt.boxplot(df['Fresh'])
plt.title('Box plot of Fresh')
```

outliers have been reduced

# K means clustering

```python
from sklearn.cluster import KMeans
```

we use elbow method to find the optimum number of clusters

```python
wcss=[]
```

```python
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df_k)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No.of clusters')
plt.ylabel('Inertia')
plt.show()

```

Optimum no.of clusters=6

# K Means model creation with no.of clusters=6

```python
model_kmeans=KMeans(n_clusters=6)
model_kmeans.fit(df_k)
labels=model_kmeans.predict(df_k)
print(labels)
```

scatter plot between Fresh and milk to see clusters

```python
fr=df_k['Fresh']
ml=df_k['Milk']
```

```python
plt.scatter(fr,ml,c=labels)
```

6 clusters can be seen with 6 different colors

To see prediction give a sample for the 6 features

```python
sample_k=[[7000,9000,7500,200,2700,1400]]
```

```python
new_lab=model_kmeans.predict(sample_k)
print(new_lab)
```

# Agglomerative Hieracrchical clustering

Normalisation


```python
from sklearn.preprocessing import Normalizer
norm=Normalizer()
df_k=norm.fit_transform(df_k)
```

```python
df_k
```

To find optimum number of cluster we use dendrogram

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked=linkage(df_k,method="ward")
plt.figure(figsize=(15,8))
dendrogram(linked,orientation="top",distance_sort="ascending",show_leaf_counts=True)
plt.hlines(2.5,0,5000,colors="r",linestyles="dashed")
plt.show()
```

optimum no.of clusters=5

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
labels_h=cluster.fit_predict(df_k)
labels_h
```

for crosschecking

```python
data=pd.DataFrame({'labels':labels_h,'Channel':channel})
data
```

```python
pd.crosstab(data['labels'],data['Channel'])
```
